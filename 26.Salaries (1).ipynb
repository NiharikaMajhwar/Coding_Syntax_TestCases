```python
import pandas as pd 
```

```python
df=pd.read_csv("C:/Users/admin/Downloads/26.Salaries.csv",encoding='ISO-8859-1')
df
```

```python
#Filter data for years 2011 and 2014
data_2011 = df[df['Year'] == 2011]
data_2014 = df[df['Year'] == 2014]

#Calculate total salary cost for each year
total_salary_2011 = data_2011['TotalPayBenefits'].sum()
total_salary_2014 = data_2014['TotalPayBenefits'].sum()

#Calculate the increase in total salary cost
increase_in_salary_cost = total_salary_2014 - total_salary_2011

print("Total salary cost increase from 2011 to 2014:", increase_in_salary_cost)
```

```python
#Find the top-earning employee across all years
top_earning_employee = df.loc[df['TotalPayBenefits'].idxmax()]

print("Top-earning employee:")
print(top_earning_employee['EmployeeName'], "with a total earning of", top_earning_employee['TotalPayBenefits'])
```
