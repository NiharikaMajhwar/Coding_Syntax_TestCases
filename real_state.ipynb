```python
import pandas as pd 
```

```python
df=pd.read_csv("C:/Users/admin/Downloads/12.real state.csv", encoding='ISO-8859-1')
df
```

```python
#City wise list of all Villas which are not less than ten thousand
villas_above_10000 = df[(df['type'] == 'Residential') & (df['price'] >= 10000) & (df['beds'] == 'Villa')]
print("City wise list of all Villas not less than $10,000:")
print(villas_above_10000[['city', 'street', 'sq__ft', 'sale_date']])
```

```python
#In GALT city, residential types with more than 800 sq__ft
galt_residential = df[(df['city'] == 'GALT') & (df['type'] == 'Residential') & (df['sq__ft'] > 800)]
galt_residential_details = galt_residential[['street', 'sq__ft', 'sale_date', 'city']]
print("\nDetails of residential types with more than 800 sq__ft in GALT city:")
print(galt_residential_details)
```

```python
#Cheapest Villa in CA: Name the city, street, and price
cheapest_villa = df[(df['type'] == 'villa') & (df['state'] == 'CA')].sort_values('price').iloc[0]
cheapest_villa_info = {
    'city': cheapest_villa['city'],
    'street': cheapest_villa['street'],
    'price': cheapest_villa['price']
}
print("\nCheapest Villa in CA:")
print(cheapest_villa_info)
```

```python
#List top 5 residencies meeting specific criteria
top_residencies = df[(df['price'] >= 60000) & (df['price'] <= 120000) &
                       (df['sq__ft'] > 1450) & (df['beds'] >= 3) & (df['baths'] >= 2)].head(5)
print("\nTop 5 residencies within budget and specific criteria:")
print(top_residencies)
```

```python
#Separate list of residential apartments with more than 2 beds and specific columns
apartments_2beds = df[(df['type'] == 'Residential') & (df['beds'] > 2)]
apartments_2beds_columns = apartments_2beds[['city', 'baths', 'sq__ft', 'price', 'type', 'beds']]
print("\nList of residential apartments with more than 2 beds:")
print(apartments_2beds_columns)
```
